Data Cleaning

```python
df["Salary Estimate"].unique()
```

Convert Salary from categorical to numerical

```python
salary = df["Salary Estimate"].apply(lambda x : x.split('(')[0])
salary
```

```python
salary = salary.apply(lambda x: x.replace("$","").replace("K",""))
salary
```

Make three different columns of min, max and avg salary

```python
df["Min Salary in K"]=salary.apply(lambda x : x.split("-")[0])
df["Max Salary in K"]=salary.apply(lambda x : x.split("-")[1])
```

```python
df
```

```python
df["Min Salary in K"].unique()
```

```python
df["Min Salary in K"].dtype
```

```python
#convert into integer 
df["Min Salary in K"]= df["Min Salary in K"].astype(int)
```

```python
df["Min Salary in K"].dtype
```

```python
df["Max Salary in K"]= df["Max Salary in K"].astype(int)
df["Max Salary in K"].dtype
```

```python
df["avg_salary"] = (df["Min Salary in K"]+df["Max Salary in K"])/2
```

```python
df["avg_salary"].dtype

```

```python
df["Company Name"].value_counts()
```

```python
#sort the company according to the rating
df[['Company Name','Rating']].sort_values(by = 'Rating', ascending=False)
```

```python
#since above we can see that the company with rating more than 0 has some issues in its name
#so discard the last 4 characters from company name
df["Company_txt"] = df.apply(lambda x : x["Company Name"] if x["Rating"]<0 else x["Company Name"][:-4], axis =1)
```
